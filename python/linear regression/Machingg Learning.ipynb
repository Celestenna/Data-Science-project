### Introduction to Machine Learning

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('SimpleLinearRegression.csv')
```

```python
df.info()
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
# correlation to check if 2 values are correlated
df.corr()
```

```python
# put on heatmap

sns.heatmap(df.corr(), annot=True, fmt='2%')

plt.show()
```

```python
sns.scatterplot(x=df['x'],y=df['y'])
plt.title('Relationship of X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
sns.regplot(x=df['x'],y=df['y'],line_kws=dict(color='y'))
plt.title('Relationship of X and Y')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# import the machine learning model

from sklearn.linear_model import LinearRegression # t make prediction
from sklearn.model_selection import train_test_split #to split
```

### Identify the dependent andindependent variable

```python
# independent feature
X = np.array(df['x']).reshape(-1,1)
```

```python
X.shape
```

```python
# dependent variable

y = df['y']
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, shuffle=False)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
# tran the model

lin_reg=LinearRegression()
```

```python
# fit in the trained data
lin_reg.fit(X_train,y_train)
```

```python
# interept
lin_reg.intercept_
```

```python
lin_reg.coef_
```

```python
# lets make a prediction
y_pred= lin_reg.predict(X_test)

y_pred
```
